```python
sr = 8000 # sample rate in Hertz.
signal, _ = librosa.load('example_speech.wav', sr=sr) # Load signal with librosa.

# trim audio
start_t, end_t = 2.5, 4.4
portion = trim_audio(signal, sr, start_t, end_t)

# Display audios
multiple_audio_widgets(sr, signal, portion, titles=["Original audio", "Audio segment"], cols=2)
```

```python
# Plot signals
times = np.arange(portion.size) / sr + start_t

axes = new_axes(1, 2) # useful to handle multiple images in matplotlib

plot_signal(signal, sr=sr, ax=axes[0], title='Original audio')
plot_signal(portion, times=times, ax=axes[1], title='Audio segment')
```

```python
# Display spectrograms.

X1 = Spectrogram(signal,  sr=sr, window_size_s=80e-3)
X2 = Spectrogram(portion, sr=sr, window_size_s=80e-3)

axes = new_axes(1, 2) # useful to handle multiple images in matplotlib

display_spectrogram(X1.spectrogram(), f_bins=X1.f_bins, t_frames=X1.t_frames, ax=axes[0], log=True, title="Original signal")
display_spectrogram(X2.spectrogram(), f_bins=X2.f_bins, t_frames=X2.t_frames + start_t, ax=axes[1], log=True, title="Trimmed signal")
```
